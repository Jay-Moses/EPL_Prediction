```python
def rolling_avarages(group, cols, new_cols):
    group = group.sort_values ("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] =rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["home_clearances", "home_corners",	"home_fouls_conceded",	"home_offsides", "home_passes",	"home_possession", "home_red_cards", "home_shots",	"home_shots_on_target",	"home_tackles",	"home_touches",	"home_yellow_cards"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_avarages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("home_team").apply(lambda x: rolling_avarages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('home_team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"]< '2020-01-01']
    test = data[data["date"]> '2020-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
